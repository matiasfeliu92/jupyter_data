```python
import pandas as pd
from datetime import datetime, timedelta
import requests
from io import StringIO
```

### URL BASE DE LA POKEAPI  https://pokeapi.co/api/v2/

```python
base_url = "https://pokeapi.co/api/v2"
```

### PETICION GET PARA OBTENER TODOS LOS POKEMONES, CON SU NOMBRE Y URL

```python
res = requests.get(f"{base_url}/pokemon/?offset=20&limit=200")
data = res.json()
len(data['results'])
```

### GUARDADO DE LOS DATOS OBTENIDOS EN UN CSV PARA SU POSTERIOR USO

```python
df_pokemons_urls = pd.json_normalize(data['results'])
df_pokemons_urls.to_csv('../CSV/pokemons_urls.csv', index=False)
df_pokemons_urls
```

### CSV GUARDADO 

```python
df_pokemons_urls = pd.read_csv('../CSV/pokemons_urls.csv')
df_pokemons_urls
```

```python
df_pokemons = pd.DataFrame()
```

```python
for url in df_pokemons_urls['url']:
    print(url)
    res = requests.get(url)
    data = res.json()
    df = pd.json_normalize(data)
    df_pokemons = pd.concat([df_pokemons, df], ignore_index = True)
```
