```python
import pandas as pd
from datetime import datetime, timedelta
import requests
from io import StringIO
```

### URL BASE DE LA POKEAPI  https://pokeapi.co/api/v2/

```python
base_url = "https://pokeapi.co/api/v2"
```

### PETICION GET A https://pokeapi.co/api/v2/pokemon/ PARA OBTENER TODOS LOS POKEMONES, CON SU NOMBRE Y URL

```python
res = requests.get(f"{base_url}/pokemon")
data = res.json()
len(data['results'])
```

### GUARDADO DE LOS DATOS OBTENIDOS EN UN CSV PARA SU POSTERIOR USO

```python
df_pokemons_urls = pd.json_normalize(data['results'])
df_pokemons_urls.to_csv('../CSV/pokemons_urls.csv', index=False)
df_pokemons_urls
```

### CSV GUARDADO 

```python
df_pokemons_urls = pd.read_csv('../CSV/pokemons_urls.csv')
df_pokemons_urls
```

```python
df_pokemons = pd.DataFrame()
df_habilities = pd.DataFrame()
df_moves = pd.DataFrame()
```

```python
for url in df_pokemons_urls['url']:
    print(url)
    res = requests.get(url)
    data = res.json()
    
    df = pd.json_normalize(data['abilities'])
    df['name'] = data['name'] 
    df_habilities = pd.concat([df_habilities, df], ignore_index = True)
    
    df = pd.json_normalize(data)
    df_pokemons = pd.concat([df, df_pokemons], ignore_index = True)
    
    df = pd.json_normalize(data['moves'])
    df['name'] = data['name'] 
    df_moves = pd.concat([df_moves, df], ignore_index = True)
```

```python
selected_columns = ['base_experience', 'height', 'id', 'is_default',  'name', 'order', 'stats', 'types', 'weight']
df_pokemons = df_pokemons[selected_columns]
df_pokemons
```

```python
df_abilities_moves = pd.merge(df_habilities, df_moves, on=['name'], how='inner')
```

```python
df_pokemons = pd.merge(df_pokemons, df_abilities_moves, on=['name'], how='inner')
df_pokemons
```
