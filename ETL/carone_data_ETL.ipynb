```python
import pandas as pd
from datetime import datetime, timedelta
import requests
from io import StringIO
from bs4 import BeautifulSoup
from lxml import html
```

```python
base_url = "https://www.carone.com.ar"
```

## ETL de los links de usados y 0km

### usando https://www.carone.com.ar/categoria-producto/0km/ para obtener la lista de autos 0km

```python
okm_model_links = []

i = 0
while True:
    res = requests.get(f"{base_url}/categoria-producto/0km/page/{i+1}/")
    print(f"{base_url}/categoria-producto/0km/page/{i+1}/")
    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.text, "html.parser")
    tree = html.fromstring(res.content)
    div_col_12_elements = tree.xpath('//div[contains(@class, "col-12")]')
    okm_page_links = [div.xpath('.//a/@href') for div in div_col_12_elements]
    okm_linkss = set([link[0] for link in okm_page_links if link != []])

    i += 1

    for link in okm_linkss:
#       print({'type': '0km','link_web': link})
      model_dict = {
          'type': '0km',
          'link_web': link
      }
      okm_model_links.append(model_dict)
```

### obtenemos los links para cada modelo 0km y los almacenamos en un dataframe

```python
df_okm_model_links = pd.DataFrame(okm_model_links)
df_okm_model_links
```

### usando https://www.carone.com.ar/categoria-producto/usados/ para obtener la lista de autos usados

```python
used_model_links = []

i = 0
while True:
    res = requests.get(f"{base_url}/categoria-producto/usados/page/{i+1}/")
    print(f"{base_url}/categoria-producto/usados/page/{i+1}/")
    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.text, "html.parser")
    tree = html.fromstring(res.content)
    div_col_12_elements = tree.xpath('//div[contains(@class, "col-12")]')
    used_page_links = [div.xpath('.//a/@href') for div in div_col_12_elements]
    used_linkss = set([link[0] for link in used_page_links if link != []])

    i += 1

    for link in used_linkss:
#       print({'type': 'usados','link_web': link})
      model_dict = {
          'type': 'usados',
          'link_web': link
      }
      used_model_links.append(model_dict)
```

### obtenemos los links para cada modelo usado y los almacenamos en un dataframe

```python
df_used_model_links = pd.DataFrame(used_model_links)
df_used_model_links
```

```python
df_model_links = pd.concat([df_okm_model_links, df_used_model_links], ignore_index=True)
df_model_links
```

```python
df_model_links.to_csv('../CSV/df_carone_model_links.csv', index=False)
```
