```python
import pandas as pd
import datetime
import requests
from io import StringIO
from bs4 import BeautifulSoup
from lxml import html
import time
```

```python
base_url = "https://datos.cultura.gob.ar/dataset/espacios-culturales-argentina-sinca"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
```

```python
response = requests.get(base_url, headers=headers)

if response.status_code != 200:
  print('url is invalid, cannot access')

soup = BeautifulSoup(response.text, "html.parser")

links_cultura = [link['href'] for link in soup.find_all('a') if 'csv' in link.get('href', '')]
print(links_cultura)
```

```python
df_bibliotecas = pd.DataFrame()
df_museos = pd.DataFrame()
```

```python
for link in links_cultura:
  try:
    if 'biblio' in link:
      print(link)
      response = requests.get(link)
      response.raise_for_status()

      csv_data = StringIO(response.text)

      df = pd.read_csv(csv_data)
      print(df.columns)
      print(len(list(df.columns)))
      df.rename(columns={
          'IdProvincia': 'id_provincia',
          'IdDepartamento': 'id_departamento',
          'Observaciones': 'observacion',
          'CategorÃ­a': 'categoria',
          'SubCategoria': 'subcategoria',
          'Provincia': 'provincia',
          'Departamento': 'departamento',
          'Localidad': 'localidad',
          'Nombre': 'nombre',
          'DirecciÃ³n': 'domicilio',
          'Piso': 'piso',
          'CP': 'cp',
          'cod_tel': 'cod_area',
          'Mail': 'mail',
          'Web': 'web',
          'InformaciÃ³n adicional': 'informacion_adicional',
          'Latitud': 'latitud',
          'Longitud': 'longitud',
          'TipoLatitudLongitud': 'tipo_latitud_longitud',
          'TemÃ¡tica': 'tematica',
          'Fuente': 'fuente',
          'aÃ±o_actualizacion': 'anio_actualizacion'
      }, inplace=True)
      df_bibliotecas = pd.concat([df_bibliotecas, df], ignore_index=True)
    elif 'museo' in link:
      print(link)
      response = requests.get(link)
      response.raise_for_status()

      csv_data = StringIO(response.text)

      df = pd.read_csv(csv_data)
      print(df.columns)
      print(len(list(df.columns)))
      df.rename(columns={
          'Cod_Loc': 'cod_localidad',
          'IdProvincia': 'id_provincia',
          'IdDepartamento': 'id_departamento',
          'Observaciones': 'observacion',
          'CP': 'cp',
          'Mail': 'mail',
          'Web': 'web',
          'Info_adicional': 'informacion_adicional',
          'Latitud': 'latitud',
          'Longitud': 'longitud',
          'TipoLatitudLongitud': 'tipo_latitud_longitud',
      }, inplace=True)
      df_museos = pd.concat([df_museos, df], ignore_index=True)
  except requests.exceptions.HTTPError as http_err:
      print(f"Error HTTP: {http_err}")
  except Exception as err:
      print(f"Otro error: {err}")
```
